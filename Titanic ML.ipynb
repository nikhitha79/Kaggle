```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/titanic/train.csv')
```

```python
df.head()
```

```python
df.info()
# Age column has some missing values. Cabin column has a lot of missing values while Embarked column is missing 2 values
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='coolwarm')
```

```python
sns.countplot(x='Survived',hue='Sex',data=df,palette='RdBu')
# 1 indicates survived and 0 indicates not survived. As expected, more number of passengers who survived were female.
```

```python
#most of the 1 class passengers were of higher age when compared to other classes. From this we can fill the missing values of age 
sns.boxplot(x='Pclass',y='Age',data=df,palette='viridis')
```

```python
def guess_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 38

        elif Pclass == 2:
            return 30

        else:
            return 25

    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass']].apply(guess_age,axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='winter')
#There are no null values in the Age column
```

```python
df.drop('Cabin',axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df['Sex']=np.where(df['Sex']=='male',0,1)
#Changes the Sex colum to integers. 0 for male and 1 for female
```

```python
df.head()
```

```python
df['Embarked']=df['Embarked'].astype('category').cat.codes
# df['Embarked'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['PassengerId','Survived','Name','Ticket'],axis=1)
```

```python
X
```

```python
y=df['Survived']
#target column
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
lgml = LogisticRegression()
lgml.fit(X_train,y_train)
```

```python
predictions = lgml.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
