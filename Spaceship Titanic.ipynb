```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train=pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
```

```python
train.info()
```

```python
train.head()
```

```python
train.drop('PassengerId',axis=1,inplace=True)
```

```python
train['CryoSleep']=np.where(train['CryoSleep']==False,0,1)
train['VIP']=np.where(train['VIP']==False,0,1)
train['Transported']=np.where(train['Transported']==False,0,1)

```

```python
train['Destination']=train['Destination'].astype('category').cat.codes
```

```python
train['HomePlanet']=train['HomePlanet'].astype('category').cat.codes
```

```python
train.head()
```

```python
train['Cabin_new']=train['Cabin'].str.split('/').str[-1]
```

```python
train['Cabin_new']=train['Cabin_new'].astype('category').cat.codes
```

```python
train.drop('Name',axis=1,inplace=True)
```

```python
train
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='VIP',y='Age',data=train,palette='winter')
```

```python
def guess_age(cols):
    Age = cols[0]
    VIP = cols[1]
    
    if pd.isnull(Age):

        if VIP == 1:
            return 35

        elif VIP == 0:
            return 27

    else:
        return Age
```

```python
train['Age'] = train[['Age','VIP']].apply(guess_age,axis=1)
```

```python
train.corr().transpose()
```

```python
train[train['Age']<=20]
```

```python
train[train['Age']>=45]
```

```python
train['VIP'].fillna(train['VIP'].median())
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
train['FoodCourt'].fillna(train['FoodCourt'].mean(),inplace=True)
```

```python
train['ShoppingMall'].fillna(train['ShoppingMall'].mean(),inplace=True)
train['Spa'].fillna(train['Spa'].median(),inplace=True)

```

```python
train['RoomService'].fillna(train['RoomService'].median(),inplace=True)
train['VRDeck'].fillna(train['VRDeck'].median(),inplace=True)


```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
```

```python
sc=StandardScaler()
```

```python
X=train.drop('Transported',axis=1)
y=train['Transported']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression(penalty='none',max_iter=1000)
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
